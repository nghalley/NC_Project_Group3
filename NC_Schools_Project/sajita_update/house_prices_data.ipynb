```python
# import libraries
import pandas as pd
from glob import glob
```

```python
# creating a list of separate csv files in a diretory
# all data files are now into list store_csv
store_csv = sorted(glob("data1/median_listingprice_*.csv"))
store_csv
```

```python
# concatinate all separate CSV files into one dataframe
combined_data = pd.concat((pd.read_csv(file).assign(filename = file)
           for file in store_csv), ignore_index = True)
combined_data
```

```python
# only taking the columns that we needed
final_df = combined_data[["regionname", "countyname", "sizerank", "2017_01",  "2017_02", "2017_03", "2017_04", "2017_05", "2017_06", "2017_07", "2017_08", "2017_09"]]
final_df
```

```python
# calculating median for the individual month
median_2017 = final_df[["2017_01", "2017_02", "2017_03", "2017_04", "2017_05", "2017_06", "2017_07", "2017_08", "2017_09"]].median(axis=1)
median_2017
```

```python
# appending the mean column into dataframe
final_df["Median_2017"] = median_2017
final_df
```

```python
# checking the columns before renaming
print(final_df.columns)
```

```python
# renaming the columns   
final_df.rename(columns = {"regionname": "Zipcode", "countyname": "County"}) 
print(final_df.columns)
```
