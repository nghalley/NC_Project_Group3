```python
import pandas as pd
from pprint import pprint
import requests
import json
```

```python
url='https://api.schooldigger.com/v1.2/schools?st=NC&city=Charlotte&page=1&perPage=50&appID=1b789824&appKey=ae269c64511c0d65f57b507e2a200ee2'
```

```python
verify = 'cacert.pem'
```

```python
schools_response = requests.get(url, verify = verify)
schools_json = schools_response.json()
```

```python
schools = pprint(schools_json)
```

```python
range(len(schools_json["schoolList"]))
```

```python
schoolList = schools_json["schoolList"]
schoolList
```

```python
school_dict = [] 

# Loop through all the cities in our list
for school in range(len(schoolList)):
       
    #response = requests.get(url, verify = verify).json() # Api call
    school_lng = schoolList[school]["address"]["latLong"]["longitude"]
    school_lat = schoolList[school]["address"]["latLong"]["latitude"]
    school_names = schoolList[school]["schoolName"]
    school_street = schoolList[school]["address"]["street"]
    school_states = schoolList[school]["address"]["state"]
    school_city = schoolList[school]["address"]["city"]
    school_county = schoolList[school]["county"]["countyName"]
    school_zip = schoolList[school]["address"]["zip"]
    school_level = schoolList[school]["schoolLevel"]
    school_year = schoolList[school]["schoolYearlyDetails"][0]["year"]
    students_number = schoolList[school]["schoolYearlyDetails"][0]["numberOfStudents"]
   
    
    school_dict.append({
        "schoolName": school_names,
        "street": school_street,
        "zip": school_zip,
        "state": school_states,
        "city": school_city,
        "countyName": school_county,
        "schoolYear" : school_year,
        "schoolLevel": school_level,
        "longitude": school_lng,
        "latitude": school_lat,
        "studentsNumber": students_number
        })    
        
charlotte_data1 = pd.DataFrame(school_dict)
charlotte_data1
```

```python
charlotte_data1.to_csv("charlotte_data1.csv", index=False)
```
